<h1>Обзор тем в Jupyter Themes</h1>

<h3><span style="color: green">Немного команд для терминала</span></h3>

```python
#Памятка
themes = ['onedork', 
          'oceans16',
          'chesterish',
          'monokai',
          'gruvboxd',
          'gruvboxl',
          'grade3',
          'solarizedl',
          'solarizedd']
```

Чтобы узнать названия всех тем, вводим в терминал команду <span style="font-family: monospace, monospace; color:#31EE78;">jt -l </span>.

```python
!jt -l
```

<div class="alert alert-block alert-warning">
        Обязательно после выполнения содержимого любой ячейки, содержащей команду для изменения графической составляющей среды следует перезапустить страницу.
</div>

Для смены темы предназначена команда <span style="font-family: monospace, monospace; color:#31EE78;">jt -t <theme> </span>. После выполнения содержимого ячейки перезагружаем страницу.

```python
!jt -t chesterish
```

Чтобы изменить отступы слева и справа от ячеек используем команду <span style="font-family: monospace, monospace; color:#31EE78;">jt -m <value> </span>.

```python
!jt -m 100
```

<h3><span style="color: green">Изменение стиля графиков Matplotlib</span></h3>

```python
import matplotlib.pyplot as plt
import numpy as np

from jupyterthemes import jtplot
```

```python
x = np.linspace(-5, 5, 1000)

y = 1 / (1 + np.exp(-x)) #sigmoid
```

```python
def plot(x, y):
    plt.clf()
    plt.figure(figsize=(14, 7))
    plt.plot(x, y, color='red', label=r'$\frac{1}{1 + e^x}$')
    plt.legend(loc='best', fontsize=30)
    plt.grid(True)
```

```python
plot(x, y)
```

```python
jtplot.style('onedork')

plot(x, y)
```

<h3><span style="color: green">Все темы в качестве изображений</span></h3>

```python
import os
from PIL import Image

fig, axes = plt.subplots(9, 1, figsize=(30, 60))
axes = axes.flat
i = 0

for theme in themes:
    file = os.path.join('themes', f'{theme}.png')
    
    axes[i].imshow(Image.open(file))
    axes[i].set_title(theme, fontsize=25)
    axes[i].axis('off')
    i += 1
```
