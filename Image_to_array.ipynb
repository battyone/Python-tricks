```python
from PIL import Image

import numpy as np
```

Not only .jpg or .png : [docs](https://pillow.readthedocs.io/en/5.1.x/handbook/image-file-formats.html)

```python
img = Image.open(r'D:\woman.jpg')
img
```

```python
def image_to_array(img):
    return np.array(img)
```

```python
img = image_to_array(img)
img
```

```python
img = Image.fromarray(img)
img
```

```python
img.save(r'D:\copied_woman.jpg')
```
