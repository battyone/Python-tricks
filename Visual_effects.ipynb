```python
from copy import copy

from PIL import Image

import numpy as np
```

```python
img = Image.open(r'D:\woman.jpg')
img
```

```python
img = np.array(img)
img
```

```python
def to_black_and_white(img):
    for i in range(224):
        for j in range(225):
            img[i, j] = np.array([img[i, j].mean()] * 3)
```

```python
to_black_and_white(img)
image = Image.fromarray(img)
image
```

```python
img = Image.open(r'D:\woman.jpg')
img = np.array(img)

img.shape
```

```python
def to_negative(img):
    for i in range(224):
        for j in range(225):
            img[i, j] = 255 - img[i, j]
                
```

```python
to_negative(img)

Image.fromarray(img)
```

```python
def reverse(img):
    return np.fliplr(img)
```

```python
img = Image.open(r'D:\woman.jpg')
img = np.array(img)

Image.fromarray(reverse(img))
```

```python
def blue(img):
    
    arr = copy(img)
    
    for i in range(224):
        for j in range(225):
            arr[i, j, 2] = 255
            
    return arr        
```

```python
Image.fromarray(blue(img))
```
