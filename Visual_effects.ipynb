```python
from PIL import Image

import numpy as np

import matplotlib.pyplot as plt
```

```python
img = Image.open(r'D:\woman.jpg')
img
```

```python
def show(img, title=None):
    plt.figure(figsize=(6, 6))
    plt.imshow(img)
    
    if title is not None:
        plt.title(title)
        
    plt.axis('off')
```

```python
def to_black(img):
    return img.convert('L')
```

```python
show(to_black(img), 'Black-white')
```

```python
img = np.array(img)
img
```

```python
img.shape
```

```python
def to_negative(img):
    return 255 - img
```

```python
show(to_negative(img), 'Negative')
```

```python
def reverse(img):
    return np.fliplr(img)
```

```python
show(reverse(img), 'Reversed')
```

```python
def blue(img):
    arr = np.copy(img)
    arr[..., 2] = 255
    return arr
```

```python
show(blue(img), 'Blue effect')
```
